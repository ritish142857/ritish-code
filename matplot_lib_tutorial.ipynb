# Matplotlib tutorial

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(5,3.5)
```

```python
#generate data for testing
x=np.arange(0,6.28,0.1)
y=np.sin(x)
y1=np.cos(x)
```

```python
plt.plot(x,y,marker='o',ms='3',alpha=0.7,ls='None',label='sine')
plt.plot(x,y1,'r*',label='cosine')
xlab=plt.xlabel(r"$\theta$(angle in radians)");xlab.set_color('blue')
ylab=plt.ylabel('y_data');ylab.set_color('blue')
title=plt.title('This is my plot')
plt.ylim()
title.set_fontweight('bold');title.set_color('blue');title.set_style('italic')
plt.legend()
```

```python
plt.plot(x,y)
#set labels 
xlab=plt.xlabel(r"$\theta$(angle in radians)")
ylab=plt.ylabel(r"sin$(\theta)$")
title=plt.title("This is a nice plot");title.set_fontweight('bold')
ylab.set_color('blue');xlab.set_color('red');xlab.set_style('italic')
```

```python
another_x=x+1
plt.plot(another_x,np.sin(another_x)*np.exp(another_x))
title=plt.title(r'sin(x)$e^x$ vs x');title.set_fontweight('bold')
```

```python
plt.plot(x,y)
x_spaced=np.arange(0,2*np.pi,0.3)
plt.plot(x_spaced,np.sin(x_spaced),marker='o',color=(1,1,0,1),linestyle='None')  #(0,1,0,1)RGB +alpha value
plt.plot(x_spaced,np.cos(x_spaced),'b+')
```

```python
plt.plot(x_spaced,np.sin(x_spaced),'ro')
xlab=plt.xlabel(r"$\theta$(in radians)");xlab.set_fontweight('bold')
title=plt.title("sine curve");title.set_fontweight('bold');title.set_color('blue')
```

```python
plt.plot(x_spaced,np.sin(x_spaced),'mo',ls='-',lw=0.5,label='sine(x)')
plt.plot(x_spaced,np.cos(x_spaced),'gs',ls='--',alpha=0.4,label='cos(x)')
xlab=plt.xlabel(r"$\theta$(angle in radians)");xlab.set_color("blue"),xlab.set_fontweight('bold')
plt.legend()
```
