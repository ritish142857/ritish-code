```python
#making graphs
# lets give python soomething to graph
import numpy as np
x=np.array([1,2,3,4,5,6,7])  # remember how to define an array
y=x+3
```

```python
#For ploting we use matplotlib library 
import matplotlib.pyplot as plt
plt.plot(x,y,'rx')
plt.title('my first graph')
plt.xlabel('Time(fortnights)')
plt.ylabel('Distance(furlongs)')
plt.xlim(0,7)
plt.ylim(0,10)
```

```python
#lets try a more complicated forms to plot
time=np.linspace(0.0,10.0,100) # we have used here the linspace() function which produces the linearly spaced numbers
height=np.exp(-time/3.0)*np.sin(time*3)
plt.plot(time,height,'m-^')
plt.plot(time,0.3*np.sin(time*3),'g-')
plt.title('damped motion')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.legend(['damped motion','constant amplitude'],loc='upper right') # remember this format for legend
```

```python
x=np.linspace(0,10,100) #to produce linearly spaced numbers
fig,ax=plt.subplots()
ax.plot(x,x**2,'r-',label='quadratic')
ax.plot(x,x**3,'m--',label='cubic')
ax.set_xlabel('x',fontsize='20')
ax.set_ylabel('$x^2$',fontsize='20')
ax.legend(fontsize='10',loc='upper left')
ax.set_xlim(0,10)
ax.set_ylim(0,200)
ax.set_title('different graph comparision')
```
